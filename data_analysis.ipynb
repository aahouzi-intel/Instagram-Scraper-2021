```
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from wordcloud import WordCloud, STOPWORDS
```

```
# Load the scraped dataset
df = pd.read_pickle("feed.pkl")
df.head()
```

```
# First, let's visualize the most common words in nike's description.

stopwords = set(STOPWORDS)
words = ""
for descrip in df['Description']: 
    tokens = descrip.split()
    tokens = [token.lower() for token in tokens]
    words += " ".join(tokens)
  
# Create the WordCloud object
wordcloud = WordCloud(width = 800, height = 800, 
                      background_color ='white', 
                      stopwords = stopwords, 
                      min_font_size = 10).generate(words) 
  
# plot the WordCloud image                        
plt.figure(figsize = (8, 8), facecolor = None)
plt.imshow(wordcloud) 
plt.axis("off") 
plt.tight_layout(pad = 0) 
  
plt.show()
```

```
# Let's see how the number of likes correlate with the number of comments
sns.set_theme()
sns.lmplot(data=df, x="Number of comments", y="Number of likes")
```
